```python

def close_position(deal_id):
    open_positions = positions_get()
    
    open_positions = open_positions[open_positions['ticket'] == deal_id]
    order_type  = list(open_positions["type"])[0]
    symbol = list(open_positions['symbol'])[0]
    volume = list(open_positions['volume'])[0]

    if(order_type == mt5.ORDER_TYPE_BUY):
        order_type = mt5.ORDER_TYPE_SELL
        price = mt5.symbol_info_tick(symbol).bid
    else:
        order_type = mt5.ORDER_TYPE_BUY
        price = mt5.symbol_info_tick(symbol).ask
	
    close_request={
        "action": mt5.TRADE_ACTION_DEAL,
        "symbol": symbol,
        "volume": float(volume),
        "type": order_type,
        "position": deal_id,
        "price": price,
        "magic": 234000,
        "comment": "Close trade",
        "type_time": mt5.ORDER_TIME_GTC,
        "type_filling": mt5.ORDER_FILLING_IOC,
    }

    result = mt5.order_send(close_request)
    
    if result.retcode != mt5.TRADE_RETCODE_DONE:
        print(now(), "Failed to close order :(")
    else:
        print (now(), "Order successfully closed!")
```

```python
t = int(time.time()) + 60*60*24
        
prices = mt5.copy_rates_from("EURUSD", mt5.TIMEFRAME_H1, t, 10000)
      
o = [x[1] for x in prices]
h = [x[2] for x in prices]
l = [x[3] for x in prices]
c = [x[4] for x in prices]
        
data = {"Time": [x[0] for x in prices], "Open":o, "High":h, "Low":l, "Close":c}
df = pd.DataFrame(data)
```

```python



class stra(Strategy):
    n1 = 200
    n3 = 50
    n2 = 20
    

    def init(self):
        close = self.data.Close
        def sma(close, n):
            return list(ta.trend.EMAIndicator(close, n).ema_indicator())
        def adx(high, low, close):
            return list(ta.trend.ADXIndicator(high, low, close).adx())
    
        self.sma1 = self.I(sma, self.data.Close.s, self.n2)
        self.sma3 = self.I(sma, self.data.Close.s, self.n3)
        self.sma2 = self.I(sma, self.data.Close.s, self.n1)
        self.adx = self.I(adx, self.data.High.s, self.data.Low.s, self.data.Close.s)

    def next(self):
        if len(self.data.Close) < 400:
            return
            
        cc = self.data.Close[-1]
        csma_s = self.sma1[-1]
        csma_m = self.sma3[-1]
        csma_l = self.sma2[-1]
            
        if len(self.orders) > 0:
            if self.position.size == 0:
                    self.orders[0].cancel()
        
        if self.adx > 30:
            if csma_s > csma_m and cc > csma_s and cc > csma_l and csma_m > csma_l:
            
                if self.position.size == 0:
                    tp = min(self.data.Low[-30:])
                    d = abs(tp-csma_s)
                    sl = csma_s + d / 1
                    risk = abs(sl - csma_s)  / csma_s * 100
                    size = int(abs(10000 * 1 / risk))
                    #print(csma_s, sl, tp)
                    self.sell(limit = csma_s, sl = sl, tp = tp, size = size)




            if csma_s < csma_m and cc < csma_s and cc < csma_l and csma_m < csma_l:
                
                if self.position.size == 0:
                    tp = max(self.data.High[-30:])
                    d = abs(tp-csma_s)
                    sl = csma_s - d / 1
                    risk = abs(sl - csma_s)  / csma_s * 100
                    size = int(abs(10000 * 1 / risk))
                    #print(size)
                    #print(csma_s, sl, tp)
                    self.buy(limit = csma_s, sl = sl, tp = tp, size = size)
                    #print(csma_s, cc)
                    
                    


bt = Backtest(df, stra,
              cash=10000, commission=0.0002,
              exclusive_orders=True, margin=1/20.0)

output = bt.run()
clear_output()
bt.plot()
```
